# Graph Theory Project

The goal of the project is to apply Floyd-Warshall algorithm in order to find the shortest paths between all pairs of vertices in a directed and weighted graph.

To do so we will define an estimation of the shortest path between two edges iteratively. Every point of the graph is considered as a potential intermediate point. Then, for each pair of edge (i,j), the algorithm check if the path between an intermediate edge k permit to obtain a shortest path than the current known distance.

```python
# Read the graph
def read_graph(filename):
    with open(f"Graphs/{filename}", "r") as f:

        nb_vertices = int(f.readline())
        nb_edges = int(f.readline())

        edges =[]

        for i in range(nb_edges):
            line = f.readline().split()
            initial = int(line[0])
            end = int(line[1])
            weight = int(line[2])

            edges.append((initial, end, weight))

    return nb_vertices, nb_edges, edges

# Print the graph
def print_graph(nb_vertices, nb_edges, edges):
    print(f"Number of vertices = {nb_vertices}")
    print(f"Number of edges = {nb_edges}")
    print("\nList of vertices:")
    for edges in edges:
        print(f"{edges[0]}->{edges[1]} (weight = {edges[2]})")
```

```python
import os
def list_graphs():
    files = os.listdir("Graphs")
    graphs = [f for f in files if f.endswith(".txt")]
    graphs.sort()
    return graphs
```

```python
#" Represent the graph in a matrix form (adjacency matrix)

def create_matrix(nb_vertices, edges):
     INF = float('inf')
     M = [[INF] * nb_vertices for _ in range(nb_vertices)]
     for i in range(nb_vertices):
         M[i][i] = 0
     for i,e,w in edges:
         M[i][e] = w

     return M
```

```python
# Print the matrix
def print_matrix(M):
    a = len(M)
    print("\n Matrix : ")
    print(f"{'':6}", end="")
    for j in range(a):
        print(f"{j:6}", end="")
    print()

    for i in range(a):
        print(f"{i:<6}", end="")  # aligné à gauche avec 6 espaces
        for j in range(a):
            if M[i][j] == float('inf'):
                print(f"{'inf':>6}", end="")
            elif M[i][j] == float('-inf'):
                print(f"{'-inf':>6}", end="")
            else:
                print(f"{M[i][j]:>6}", end="")
        print()

    return
```
