```python
import os


# ===================================
# Fonctions to read the Graphs
# ===================================

# Read the graph
def read_graph(filename):
    with open(f"Graphs/{filename}", "r") as f:

        nb_vertices = int(f.readline())
        nb_edges = int(f.readline())

        edges =[]

        for i in range(nb_edges):
            line = f.readline().split()
            initial = int(line[0])
            end = int(line[1])
            weight = int(line[2])

            edges.append((initial, end, weight))

    return nb_vertices, nb_edges, edges

# Print the graph
def print_graph(nb_vertices, nb_edges, edges):
    print(f"Number of vertices = {nb_vertices}")
    print(f"Number of edges = {nb_edges}")
    print("\nList of vertices:")
    for edges in edges:
        print(f"{edges[0]}->{edges[1]} (weight = {edges[2]})")
```

```python
import os
def list_graphs():
    files = os.listdir("Graphs")
    graphs = [f for f in files if f.endswith(".txt")]
    graphs.sort()
    return graphs
```

```python
def main():
    print("=========Main meun===========")
    graphs = list_graphs()

    if len(graphs) == 0:
        print("No graphs found")
        return

    print("Graphs available : ")
    for i, name in enumerate(graphs):
        print(f"{i+1}. {name}")

    print(f"0. Exit")

    choice = int(input("Enter your choice : "))
    if choice == 0:
        print("Exit")
        return

    if choice < 1 or choice > len(graphs):
        print("Invalid choice")
        return

    name_file = graphs[choice-1]
    nb_vertices, nb_edges, edges = read_graph(name_file)
    print_graph(nb_vertices, nb_edges, edges)

main()
```
